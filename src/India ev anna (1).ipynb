```python
# Row-wise total for each state
daf['total'] = daf['2022 - Accidents'] + daf['2023 - Accidents'] + daf['2024 - Accidents']
daf['total']
```

## Trend Analysis

```python

# Calculate increase in accidents
daf["Accident Increase"] = daf["2024 - Accidents"] - daf["2022 - Accidents"]

# Filter only states with a positive increase
increase_df = daf[daf["Accident Increase"] > 0]

# Sort by highest increase
increase_df = increase_df.sort_values(by="Accident Increase", ascending=False)

# Top 10 states with highest increase
top10_increase = increase_df.head(10)

# Plot
plt.figure(figsize=(12, 6))
plt.barh(top10_increase["State/UT"], top10_increase["Accident Increase"], color="crimson")
plt.xlabel("Increase in Accidents (2022 to 2024)")
plt.ylabel("State/UT")
plt.title("Top 10 States with Highest Increase in Accidents (2022 to 2024)")
plt.show()
```

```python
# Print the table
print("Top 10 States with Highest Increase in Accidents (2022–2024):")
print(top10_increase[["State/UT", "2022 - Accidents", "2024 - Accidents", "Accident Increase"]])
```

```python
#Which state has the most consistent number of accidents across the years?
# Calculate standard deviation of accidents across the 3 years
daf["Accident_STD"] = daf[["2022 - Accidents", "2023 - Accidents", "2024 - Accidents"]].std(axis=1)
# Find the state with the **lowest standard deviation** (most consistent)
most_consistent_state = daf.loc[daf["Accident_STD"].idxmin()]
print("Most Consistent State in Terms of Accidents:")
print(most_consistent_state[["State/UT", "2022 - Accidents", "2023 - Accidents", "2024 - Accidents", "Accident_STD"]])
 # Visualize consistency across all states
sorted_df = daf.sort_values(by="Accident_STD")
plt.figure(figsize=(14, 6))
plt.barh(sorted_df["State/UT"], sorted_df["Accident_STD"], color="mediumseagreen")
plt.xlabel("Standard Deviation of Accidents")
plt.title("Consistency of Accident Counts (Lower = More Consistent)")
plt.show()
```

```python
#Are accident numbers increasing overall across India over the years?

total_accidents = {
    "2022": daf["2022 - Accidents"].sum(),
    "2023": daf["2023 - Accidents"].sum(),
    "2024": daf["2024 - Accidents"].sum()
}
accident_trend = pd.DataFrame(list(total_accidents.items()), columns=["Year", "Total Accidents"])
print("Total Accidents Over the Years:")
print(accident_trend)
```

```python
#Which states have the highest accident fatality rate (deaths per accident) over the years?
#Calculate fatality rates for each year
daf["2022 Fatality Rate"] = daf["2022 - Deaths"] / daf["2022 - Accidents"].replace(0, pd.NA)
daf["2023 Fatality Rate"] = daf["2023 - Deaths"] / daf["2023 - Accidents"].replace(0, pd.NA)
daf["2024 Fatality Rate"] = daf["2024 - Deaths"] / daf["2024 - Accidents"].replace(0, pd.NA)
daf["Average Fatality Rate"] = daf[["2022 Fatality Rate", "2023 Fatality Rate", "2024 Fatality Rate"]].mean(axis=1)
top_fatality_states = daf.sort_values(by="Average Fatality Rate", ascending=False).head(10)
print(top_fatality_states[["State/UT", "Average Fatality Rate"]])
```

```python
#Which states have the highest injury rate (injured per accident)?
daf['Injury Rate'] = daf['tot_inju'] / daf['total']
daf[['State/UT', 'Injury Rate']].value_counts().sort_values(by = 'Injury Rate')
```

```python
# Are deaths increasing in states where accidents are rising? Is there a correlation?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Step 1: Calculate changes
daf["Accident Change"] = daf["2024 - Accidents"] - daf["2022 - Accidents"]
daf["Death Change"] = daf["2024 - Deaths"] - daf["2022 - Deaths"]
# Step 2: Correlation calculation
correlation, p_value = pearsonr(daf["Accident Change"], daf["Death Change"])
print(f"Pearson Correlation between accident change and death change: {correlation:.2f}")
print(f"P-value: {p_value:.4f} {'(Significant)' if p_value < 0.05 else '(Not Significant)'}")
# Step 3: Visualization
correlation_matrix = daf[[
    "2022 - Accidents", "2023 - Accidents", "2024 - Accidents",
    "2022 - Deaths", "2023 - Deaths", "2024 - Deaths",
    "2022 - Injured", "2023 - Injured", "2024 - Injured"
]].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix: Accidents, Deaths, Injuries (2022-2024)")
plt.tight_layout()
plt.show()
```

```python
#Which state reported the highest number of accidents in 2024?
# Sort the DataFrame by 2024 accident count in descending order
top_state = daf.sort_values(by="2024 - Accidents", ascending=False).head(1)

# Display the result
print("State with the highest number of accidents in 2024:")
display(top_state[["State/UT", "2024 - Accidents"]])
```
